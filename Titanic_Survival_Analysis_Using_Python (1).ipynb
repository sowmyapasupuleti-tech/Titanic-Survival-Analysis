# Titanic Survival Analysis Using Python

## Objective
The objective of this project is to analyze the Titanic dataset, perform data cleaning, exploratory data analysis (EDA), and identify factors that influenced passenger survival.

## Tools Used
- Python
- Pandas
- Matplotlib
- Google Colab

## Dataset
Titanic Passenger Dataset

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

df['Survived'].value_counts().plot(kind='bar')

plt.title("Survival Distribution")

plt.show()
```

```python
pd.crosstab(df['Sex'], df['Survived'])
```

```python
pd.crosstab(df['Sex'], df['Survived']).plot(kind='bar')

plt.title("Survival by Gender")

plt.xlabel("Gender")

plt.ylabel("Number of Passengers")

plt.show()
```

```python
pd.crosstab(df['Pclass'], df['Survived'])
```

```python
pd.crosstab(df['Pclass'], df['Survived']).plot(kind='bar')

plt.title("Survival by Passenger Class")

plt.xlabel("Passenger Class")

plt.ylabel("Number of Passengers")

plt.show()
```

```python
pd.crosstab(df['Pclass'], df['Survived'])
```

# Key Findings

1. The dataset contained missing values in Age, Cabin, and Embarked columns.

2. Missing values were handled using median and mode imputation techniques.

3. The Cabin column was removed due to a large number of missing values.

4. Female passengers had a significantly higher survival rate than male passengers.

5. Passenger class influenced survival chances, with higher-class passengers generally showing better survival outcomes.

6. Data visualization helped identify important survival patterns.

# Conclusion

The analysis revealed that gender and passenger class significantly influenced survival outcomes. Female passengers had higher survival rates, and higher-class passengers generally had better chances of survival. Data cleaning and visualization techniques helped uncover these insights.
